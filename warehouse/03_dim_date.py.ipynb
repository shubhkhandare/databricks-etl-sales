```python
from pyspark.sql.functions import col, year, month, dayofmonth, quarter, date_format
from pyspark.sql.types import DateType
import pandas as pd
```

```python
from pyspark.sql.functions import to_date
import pandas as pd
```

```python
start_date = "2020-01-01"
end_date   = "2030-12-31"

dates = pd.date_range(start=start_date, end=end_date)

date_df = spark.createDataFrame(
    [(d.strftime("%Y-%m-%d"),) for d in dates],
    ["date_str"]
).withColumn("date", to_date("date_str"))
```

```python
date_df.printSchema()
```

```python
date_df.show(5)
```

```python
dim_date_df = (
    date_df
    .withColumn("date_id", date_format(col("date"), "yyyyMMdd").cast("int"))
    .withColumn("year", year(col("date")))
    .withColumn("month", month(col("date")))
    .withColumn("day", dayofmonth(col("date")))
    .withColumn("quarter", quarter(col("date")))
    .withColumn("day_name", date_format(col("date"), "EEEE"))
    .withColumn("month_name", date_format(col("date"), "MMMM"))
)
```

```python
dim_date_df.display()
```

```python
dim_date_path = "/Volumes/workspace/default/etl_sales/warehouse/dim_date"

dim_date_df.write \
    .format("delta") \
    .mode("overwrite") \
    .save(dim_date_path)
```
